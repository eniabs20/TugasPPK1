Contoh lain lagi dari proses data selection: Misal kita ingin membuat linearly spaced array mulai angka -2*PI sampai 2*PI.


```python
import numpy as np
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x**2)
import matplotlib.pyplot as plt
plt.plot(x, y) 
plt.show()
```

```python
mask = y>=0
plt.plot(x, y) 
plt.plot(x[mask], y[mask], 's') 
plt.show()
```

Contoh selanjutnya menggambar plot untuk fungsi trigonometry:


```python
from numpy import pi
from numpy import sin
from numpy import cos
x = linspace(-2*pi, 2*pi, 200)
y1 = sin(x)
y2 = cos(x)
plt.plot(x, y1) 
plt.plot(x, y2) 
plt.show()
```

## Menyunting Properti Plot

```python
plt.plot(x, y1, 'r') 
plt.plot(x, y2, 'g') 
plt.grid() 
plt.xlim(-2*pi, 2*pi) 
plt.ylim(-1.1, 1.1)
plt.show()
```

## Memberi Label pada Plot

```python
plt.plot(x, y1, 'r') 
plt.plot(x, y2, 'g')
plt.grid() 
plt.xlim(-2*pi, 2*pi) 
plt.ylim(-1.1, 1.1)
plt.xlabel('x') 
plt.ylabel('sin and cos functions') 
plt.title('A plot of sin and cos functions') 
plt.show()
```

## Menggunakan Syntax Latex pada Plot

```python
 x = linspace(-2*pi, 2*pi, 200) 
 y1 = sin(x**3) 
 plt.plot(x, y1, 'r') 
 plt.grid() 
 plt.xlim(-2*pi, 2*pi)  
 plt.ylim(-1.1, 1.1)
 plt.xlabel('x')  
 plt.ylabel('$y = sin(x^3)$') # <= syntax latex 
 plt.title('A plot of a trig function') 
 plt.show()
```
