# Latihan Matplotlib

(1) Cari bagaimana cara menampilkan Judul (title) dari grafik

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5,5,100)
y = 2*x+1
plt.plot(x, y, '-r')
#Menampilkan judul (title) dari grafik
plt.title('Grafik dari y = 2x+1')
plt.grid()
plt.show()
```

(2) Cari bagaimana cara memberikan label pada sumbu x dan sumbu y

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5,5,100)
y = 2*x+1
plt.plot(x, y, '-r')
plt.title('Grafik dari y = 2x+1')
#Memberikan label pada sumbu x dan sumbu y
plt.xlabel('Sumbu X', color='#1C2833')
plt.ylabel('Sumbu Y', color='#1C2833')
plt.grid()
plt.show()
```

(3) Cari bagaimana cara menambah teks dalam grafik

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5,5,100)
y = 2*x+1
plt.plot(x, y, '-r', label='y = 2x+1')
plt.title('Grafik dari y = 2x+1')
plt.xlabel('Sumbu X', color='#1C2833')
plt.ylabel('Sumbu Y', color='#1C2833')
#Menambah teks dalam grafik
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
